# Evaluation 

In the evaluation, all of the models will be evaluated by getting the metrics: precision, recall and F1-score.

In order to achieve this, the above process will be repeated for the other models and the extracted requirements will be saved for each model. Finally, it is possible to evaluate.

The process to extract requirements for the eight models can be visualized in the "Extraction_and_saving_requirements.ipynb" file.

```python
dataset_label = ['test_label_eBay.txt', 'test_label_Evernote.txt', 'test_label_Facebook.txt', 'test_label_Netflix.txt', 
                 'test_label_PhotoEditor.txt', 'test_label_Spotify.txt', 'test_label_Twitter.txt', 'test_label_WhatsApp.txt']
dataset_extracted = ['CV_0_extracted_reqs.txt', 'CV_1_extracted_reqs.txt', 'CV_2_extracted_reqs.txt', 'CV_3_extracted_reqs.txt', 
                     'CV_4_extracted_reqs.txt', 'CV_5_extracted_reqs.txt', 'CV_6_extracted_reqs.txt', 'CV_7_extracted_reqs.txt']                 
```

```python
from evaluation import feature_matching, metrics, f1_measure

for i in range(0, len(dataset_label)): 
    f = open(dataset_label[i], 'r')
    test_label = f.readlines()
    f.close()

    f2 = open(dataset_extracted[i], 'r')
    test_extracted = f2.readlines()
    f2.close()

    features_labeled = list()
    for label in test_label:
        feature = label.strip()
        features_labeled.append(feature)
    
    features_extracted = list()
    for extracted in test_extracted: 
        feature = extracted.strip()
        features_extracted.append(feature)

    print('# Model ' + str(i))
    for n in range(0, 3): 
        print('Precision, Recall and F1-score for n = ' + str(n) + ': ',  f1_measure(features_extracted, features_labeled, n))
```
